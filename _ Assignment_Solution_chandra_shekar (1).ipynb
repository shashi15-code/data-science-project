SECTION_2 : ** Data Analysis **

```python
import sqlite3

# connecting to an SQLite database
conn = sqlite3.connect('Housing_data.db')

# Saving 'property_details' DataFrame as a table in the database
property_details.to_sql('property_details', conn, if_exists='replace', index=False)

# Saving 'property_price_details' DataFrame as a table in the database
property_price_details.to_sql('property_price_details', conn, if_exists='replace', index=False)

```

```python
from sqlalchemy import create_engine,text
import pandas as pd

# Create a database connection
engine = create_engine('sqlite:///Housing_data.db')

# Executing sql query and storing it in dataframe
# question 1 : Retrieve properties that have a price greater than 1 million and are located in "Estados Unidos" (l1).

problem_1 ="SELECT pd.id, ppd.property_type, pd.l1, ppd.price FROM property_details pd INNER JOIN property_price_details ppd ON pd.id = ppd.id WHERE pd.l1 = 'Estados Unidos' AND ppd.price > 1000000;"

conn_sql1 = pd.DataFrame(engine.connect().execute(text(problem_1)))

conn_sql1
```

```python
# question 2 : Categorize properties based on their surface area as 'Small' if it's less than 50 square meters, 'Medium' if it's
#between 50 and 100 square meters, and 'Large' if it's greater than 100 square meters:

problem_2 = """SELECT *,
  CASE
    WHEN surface_total < 50 THEN 'Small'
    WHEN surface_total >= 50 AND surface_total <= 100 THEN 'Medium'
    WHEN surface_total > 100 THEN 'Large'
    ELSE 'Unknown'
  END AS property_size
FROM property_details; """

conn_sql2 =  pd.DataFrame(engine.connect().execute(text(problem_2)))
conn_sql2
```

```python
# question 3 : List all properties (id) in the "Belgrano" neighborhood (l3) that have the same number of bedrooms and bathrooms as another property in the dataset:


problem_3 = """select p1.id,p1.l3,p1.bedrooms,p2.bedrooms,p1.bathrooms,p2.bathrooms from property_details p1 inner join property_details p2
on p1.id = p2.id and p1.bathrooms = p2.bathrooms and p1.bedrooms = p2.bedrooms
where p1.l3 = 'Belgrano' and p2.l3 = 'Belgrano' """

conn_sql3 =  pd.DataFrame(engine.connect().execute(text(problem_3)))

conn_sql3
```

```python
# question 4 : Calculate the average price per square meter (price / surface_total) for each property type (property_type) in
#the "Belgrano" neighborhood (l3)

problem_4  = """select (b.price/a.surface_total)as price_per_square_meter,b.property_type,a.l3 from property_details a
join property_price_details b on a.id = b.id where a.l3 = 'Belgrano' group by b.property_type """

conn_sql4 =  pd.DataFrame(engine.connect().execute(text(problem_4)))

conn_sql4
```

```python
# question 5: Calculate the cumulative price for each property type, ordered by the creation date
problem_5 = """SELECT
    pd.id ,
    ppd.property_type,
    pd.created_on,
    ppd.price,
    SUM(ppd.price) OVER (PARTITION BY ppd.property_type ORDER BY pd.created_on) AS cumulative_price
FROM Property_Details pd
INNER JOIN Property_Price_Details ppd ON pd.id = ppd.id
ORDER BY ppd.property_type, pd.created_on;
 """

conn_sql5 =  pd.DataFrame(engine.connect().execute(text(problem_5)))
conn_sql5
```

```python
#Identify properties that have a higher price than the average price of properties with the same number of bedrooms and bathrooms
problem_6 = """SELECT
    pd.id,
    pd.bedrooms,
    pd.bathrooms,
    ppd.price
FROM
    property_details pd
JOIN
    property_price_details ppd
ON
    pd.id = ppd.id
WHERE
    ppd.price > (
        SELECT
            AVG(ppd2.price)
        FROM
            property_details pd2
        JOIN
            property_price_details ppd2
        ON
            pd2.id = ppd2.id
        WHERE
            pd2.bedrooms = pd.bedrooms
        AND
            pd2.bathrooms = pd.bathrooms
    ); """

conn_sql6 = pd.DataFrame(engine.connect().execute(text(problem_6)))
conn_sql6
```

```python
# question 7 : Identify the 10 locations (l3) with the highest total surface area (sum of surface_total) of properties listed for
#sale (operation_type = 'Venta'):

problem_7 = """SELECT pd.l3, SUM(pd.surface_total) AS total_surface_area
FROM Property_Details pd
INNER JOIN Property_Price_Details ppd ON pd.id = ppd.id
WHERE ppd.operation_type = 'Venta'
GROUP BY pd.l3
ORDER BY total_surface_area DESC
LIMIT 10; """

conn_sql7 = pd.DataFrame(engine.connect().execute(text(problem_7)))
conn_sql7
```

```python
# question 8 : Find the top 5 most expensive properties (based on price) in the "Palermo" neighborhood (l3) that were listed
#in August 2020:
problem_8 = """SELECT pd.id , ppd.price, pd.l3, pd.created_on
FROM property_details pd
INNER JOIN property_price_details ppd ON pd.id = ppd.id
WHERE pd.l3 = 'Palermo'
    AND ppd.operation_type = 'Venta'
    AND DATE(pd.created_on) BETWEEN '2020-08-01' AND '2020-08-31'
ORDER BY ppd.price DESC
LIMIT 5;

"""
conn_sql8 = pd.DataFrame(engine.connect().execute(text(problem_8)))
conn_sql8
```

```python
# question 9 : Find the top 3 properties with the highest price per square meter (price divided by surface area) within each
#property type.

problem_9 = """WITH RankedProperties AS (
    SELECT
        pd.id ,
        ppd.property_type,
        ppd.price,
        pd.surface_total,
        pd.surface_covered,
        ROW_NUMBER() OVER (PARTITION BY ppd.property_type ORDER BY ppd.price / pd.surface_total DESC) AS rn
    FROM Property_Details pd
    INNER JOIN Property_Price_Details ppd ON pd.id = ppd.id
    WHERE pd.surface_total > 0
)
SELECT id, property_type, price, surface_total, surface_covered
FROM RankedProperties
WHERE rn <= 3;
"""
conn_sql9 =  pd.DataFrame(engine.connect().execute(text(problem_9)))
conn_sql9
```

```python
# question 10 : Find the top 3 locations (l1, l2, l3) with the highest average price per square meter (price / surface_total) for
#properties listed for sale (operation_type = 'Venta') in the year 2020. Exclude locations with fewer than 10
#properties listed for sale in 2020 from the results.
problem_10 = """WITH LocationAverages AS (
    SELECT
        pd.l1,
        pd.l2,
        pd.l3,
        AVG(ppd.price / pd.surface_total) AS avg_price_per_sqm,
        COUNT(*) AS property_count
    FROM Property_Details pd
    INNER JOIN Property_Price_Details ppd ON pd.id = ppd.id
    WHERE ppd.operation_type = 'Venta'
      AND DATE(pd.created_on) BETWEEN '2020-08-01' AND '2020-08-31'
    GROUP BY pd.l1, pd.l2, pd.l3
    HAVING COUNT(*) >= 10
),
RankedLocations AS (
    SELECT
        l1,
        l2,
        l3,
        avg_price_per_sqm,
        ROW_NUMBER() OVER (ORDER BY avg_price_per_sqm DESC) AS rn
    FROM LocationAverages
)
SELECT l1, l2, l3, avg_price_per_sqm
FROM RankedLocations
WHERE rn <= 3;
"""
conn_sql10 =  pd.DataFrame(engine.connect().execute(text(problem_10)))
conn_sql10
```
